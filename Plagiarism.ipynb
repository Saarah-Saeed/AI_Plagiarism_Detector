```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from nltk.tokenize import sent_tokenize
import numpy as np

# Load model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Sample texts
text1 = """Artificial Intelligence is transforming the world. It helps automate tasks and improve efficiency."""

text2 = """AI is changing the globe. It automates processes and increases productivity."""

# Tokenize sentences
sentences1 = sent_tokenize(text1)
sentences2 = sent_tokenize(text2)

# Generate embeddings
embeddings1 = model.encode(sentences1)
embeddings2 = model.encode(sentences2)

# Compute similarity matrix
similarity_matrix = cosine_similarity(embeddings1, embeddings2)

# Parameters
threshold = 0.7
top_k = 2

matches = []

print("\n📊 Top Matches:\n")

# Find best matches
for i, sent1 in enumerate(sentences1):
    sim_scores = similarity_matrix[i]

    # Get top-k indices
    top_indices = np.argsort(sim_scores)[-top_k:][::-1]

    for j in top_indices:
        score = sim_scores[j]

        if score >= threshold:
            matches.append((sent1, sentences2[j], score))

            print(f"🔹 Text1: {sent1}")
            print(f"🔸 Text2: {sentences2[j]}")
            print(f"✅ Similarity: {score:.2f}")
            print("-" * 50)

# Overall similarity score
if matches:
    avg_score = np.mean([m[2] for m in matches])
else:
    avg_score = 0

# ✅ FIRST define risk
if avg_score < 0.4:
    risk = "Safe ✅"
elif avg_score < 0.7:
    risk = "Moderate ⚠️"
else:
    risk = "High Risk 🚨"

# ✅ THEN print
print("\n" + "="*60)
print(f"📈 Overall Similarity Score: {avg_score*100:.2f}%")
print(f"⚠️ Plagiarism Risk Level: {risk}")
print("="*60)
```

STEP 5: Writing Style Consistency Analyzer

```python
import numpy as np
import re

# Function to calculate average sentence length
def avg_sentence_length(sentences):
    lengths = [len(sentence.split()) for sentence in sentences]
    return np.mean(lengths)

# Function to calculate vocabulary complexity (avg word length)
def avg_word_length(text):
    words = re.findall(r'\w+', text)
    word_lengths = [len(word) for word in words]
    return np.mean(word_lengths)

# Simple readability approximation (not exact Flesch, but enough for project)
def readability_score(text):
    sentences = sent_tokenize(text)
    words = re.findall(r'\w+', text)

    num_sentences = len(sentences)
    num_words = len(words)
    num_chars = sum(len(word) for word in words)

    if num_sentences == 0 or num_words == 0:
        return 0

    avg_sentence_len = num_words / num_sentences
    avg_word_len = num_chars / num_words

    # simplified readability formula
    score = 100 - (1.5 * avg_sentence_len) - (5 * avg_word_len)
    return score

# --- Analyze Style for Both Texts ---

style_features = {
    "Text1": {
        "avg_sentence_length": avg_sentence_length(sentences1),
        "avg_word_length": avg_word_length(text1),
        "readability": readability_score(text1)
    },
    "Text2": {
        "avg_sentence_length": avg_sentence_length(sentences2),
        "avg_word_length": avg_word_length(text2),
        "readability": readability_score(text2)
    }
}

print("\n🧠 Style Analysis:\n")

for text, features in style_features.items():
    print(f"{text}:")
    print(f"  Avg Sentence Length: {features['avg_sentence_length']:.2f}")
    print(f"  Avg Word Length: {features['avg_word_length']:.2f}")
    print(f"  Readability Score: {features['readability']:.2f}")
    print()

# --- Style Consistency Score ---
diff_sentence = abs(style_features["Text1"]["avg_sentence_length"] - style_features["Text2"]["avg_sentence_length"])
diff_word = abs(style_features["Text1"]["avg_word_length"] - style_features["Text2"]["avg_word_length"])
diff_readability = abs(style_features["Text1"]["readability"] - style_features["Text2"]["readability"])

# Normalize differences
style_score = 100 - (diff_sentence * 2 + diff_word * 5 + diff_readability * 0.5)

# Clamp score between 0 and 100
style_score = max(0, min(100, style_score))

print("🎯 Style Consistency Score: {:.2f}%".format(style_score))

# Interpretation
if style_score > 75:
    print("✅ Writing style is consistent")
elif style_score > 50:
    print("⚠️ Minor style variation detected")
else:
    print("🚨 Significant style inconsistency detected")
```

```python
def generate_report(matches, avg_score, risk, style_score, tfidf_similarity):

    print("\n" + "="*70)
    print("📄 AI-BASED PLAGIARISM DETECTION REPORT")
    print("="*70)

    # Overall Scores
    print(f"\n📈 Semantic Similarity Score: {avg_score*100:.2f}%")
    print(f"⚠️ Plagiarism Risk Level: {risk}")
    print(f"🎯 Style Consistency Score: {style_score:.2f}%")

    print("\n" + "-"*70)

    # Sentence Matches
    print("\n🔍 Top Matching Sentences:\n")

    if not matches:
        print("No significant matches found.")
    else:
        for i, (s1, s2, score) in enumerate(matches, 1):
            print(f"{i}.")
            print(f"   🔹 Text1: {s1}")
            print(f"   🔸 Text2: {s2}")
            print(f"   ✅ Similarity: {score:.2f}")
            print()

    print("-"*70)

    # Comparison Section
    print("\n📊 Method Comparison:\n")
    print(f"🔹 TF-IDF Score: {tfidf_similarity*100:.2f}%")
    print(f"🔸 Semantic Score: {avg_score*100:.2f}%")

    if avg_score > tfidf_similarity:
        print("✅ Semantic method is more effective for paraphrase detection.")
    else:
        print("⚠️ Traditional method performed similarly or better.")

    print("="*70)
```

```python
generate_report(matches, avg_score, risk, style_score, tfidf_similarity)
```
